Task 1: Working with Series
1. Create a Pandas Series from the following list: 
data = [25, 30, 35, 40, 45]
2. Write a Python program to: 
o Assign custom indices ['A', 'B', 'C', 'D', 'E'] to the Series.
o Display the first three elements of the Series.
o Calculate the mean, median, and standard deviation of the 
Series.


```python
data = [25, 30, 35, 40, 45]
series = pd.Series(data)
print(series, type(series))

series = pd.Series(data, index=['A', 'B', 'C', 'D', 'E'])
print(series, type(series))
print(series[:3])

print("Mean : ", series.mean(numeric_only = True))
print("Median : ", series.median(numeric_only = True))
print("Standard Deviation: ", series.std(numeric_only = True))
```

Task 2: Creating and Inspecting DataFrames
1. Create a Pandas DataFrame with the following data:

Name Age Gender Marks

Alice 20 Female 85

2. Write a Python program to:

    o Display the first two rows of the DataFrame.

    o Display the column names, data types, and summary statistics.

and False otherwise.

```python
data = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [20, 22, 19, 21, 20],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'Marks': [85, 78, 92, 74, 88]
}

df = pd.DataFrame(data)

print("First 2 Rows :")
print(df.head(2))

print('-'*30)
print("Column Names : ")
print(df.columns.tolist())

print('-'*30)
print("Data Types : ")
print(df.dtypes)

print('-'*30)
print("Summary Statistics : ")
print(df.describe())

 
print('-'*30)
df['Passed'] = df['Marks'] >= 80

print('-'*30)
print("DataFrame with 'Passed' Column:")
print(df)
```

Task 3: Data Selection and Filtering
1. Using the DataFrame from Task 2, write a Python program to: 

o Select and display the Name and Marks columns.

o Filter and display records where Marks > 80.

o Display the record of the student with the highest marks.


```python
data = {
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve'],
    'Age': [20, 22, 19, 21, 20],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female'],
    'Marks': [85, 78, 92, 74, 88]
}

df = pd.DataFrame(data) 

df['Passed'] = df['Marks'] >= 80
print(df)
print('-'*30)
print("Name n Marks column :")
print(df[['Name','Marks']])
print('-'*30)

 
print("Students with Marks > 80 :")
print(df[df['Marks'] > 80])
print('-'*30)



max_marks = df['Marks'].max()
tooper = df[df['Marks'] == max_marks]

print("Student with the Highest Marks:")
print(tooper)

print('-'*30)
print(df)
```

Task 4: Handling Missing Data
1. Modify the DataFrame from Task 2 by introducing missing values: 
2. df.loc[1, 'Marks'] = None
3. df.loc[4, 'Age'] = None
4. Write a Python program to: 

o Identify missing values in the DataFrame.

o Fill missing values in the Marks column with the column's 
mean.

o Drop rows where the Age column has missing values.

```python
df.loc[1,'Marks'] = None  
df.loc[4,'Age'] = None  
print(" DataFrame with Missing Values:")
print(df)
print('-'*30)
 
print("Missing Values in DataFrame :")
print(df.isnull())
print('-'*30)
 
mean_marks = df['Marks'].mean()
df['Marks'].fillna(mean_marks, inplace=True)

print(f"Filled missing 'Marks' with mean ({mean_marks:.2f}):")
print(df)
print('-'*30)
df.dropna(subset=['Age'], inplace=True)

print("Dropping rows with missing Age :")
print(df)
print('-'*30)
```

Task 5: Grouping and Aggregation
1. Using the DataFrame from Task 2, write a Python program to: 

        o Group the data by Gender and calculate the mean age and 
marks for each gender.

        o Count the number of students in each gender group.


```python
group_mean = df.groupby('Gender')[['Age','Marks']].mean()
print("Mean Age and Marks by Gender:")
print(group_mean)
print('-'*30)
 
gender_count = df['Gender'].value_counts()
print("Count of Students by Gender:")
print(gender_count)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('titanic')
print("titanic data grabbed")

print('-' * 30)
print("data:")
print(data.head())

print('-' * 30)
print("stats:")
print(data.describe(include='all'))

print('-' * 30)
print(" ")
print(data.isnull().sum())

print('-' * 30)
plt.figure(figsize=(8, 5))
sns.boxplot(data=data, x='class', y='age', palette='Set2')
plt.title('how age spreads across classes')
plt.show()

quick = data[['survived', 'age', 'fare', 'class']].dropna()
sns.pairplot(quick, hue='class')
plt.suptitle("some quick plots", y=1.02)
plt.show()

nums = data.select_dtypes(include='number')
plt.figure(figsize=(6, 4))
sns.heatmap(nums.corr(), annot=True, cmap="coolwarm")
plt.title("correlation vibes")
plt.show()
```
